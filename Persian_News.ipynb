```python
data.drop(labels = ['title', 'abstract', 'time'] , axis = 1 , inplace = True)
data.head()
```

```python
x = data.drop(labels = ['category'] , axis = 1)
x = x['body']
x
```

```python
y = data.drop(labels = ['body'] , axis = 1)
y
```

```python
category_dict = y['category'].drop_duplicates().reset_index(drop = True).to_dict()
category_dict
```

```python
category_to_id = {v: k for k, v in category_dict.items()}
y['category_id'] = y['category'].map(category_to_id)
y
```

```python
y.drop(labels = ['category'] , axis = 1 , inplace = True)
y
```

```python
!wget https://raw.githubusercontent.com/stopwords-iso/stopwords-fa/master/stopwords-fa.txt -O stopwords-fa.txt

with open("stopwords-fa.txt", "r", encoding="utf-8") as f:
    stop_words = set(f.read().splitlines())

print(len(stop_words))
# print(list(stop_words)[:30])
```

```python
normalizer = parsivar.Normalizer()
tokenizer = parsivar.Tokenizer()
stop_words = set(stop_words)
punc = set(string.punctuation + '،؛؟')
stemmer = parsivar.FindStems()

def clean_text(text) :

    if pd.isna(text) :
        return ""

    text = normalizer.normalize(text)
    tokens = tokenizer.tokenize_words(text)

    filtered_tokens = []
    for token in tokens :
        if token not in stop_words and token not in punc and not token.isdigit() :
            filtered_tokens.append(stemmer.convert_to_stem(token))

    text = ' '.join(filtered_tokens)
    return text
```
